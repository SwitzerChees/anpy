```python
import matplotlib.pyplot as plt
import numpy as np

sigmoid = lambda x: 1 / (1 + np.e ** x)

x = np.arange(-10,10,0.1)
y = [sigmoid(s) for s in y]

plt.plot(x, y)
plt.show()
```

```python
def area (f, a, b, n):
      step = (b - a) / n
      sum = 0
      for i in range(n):
          sum += f(a + step/2 + i * step)
      return sum * step


print(area(sigmoid, -9, 10, 256))
```
