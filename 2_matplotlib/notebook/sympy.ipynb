### Stetigkeit und Brücke zur numerischen Integration



```python
# konvergierende Reihe

x = np.arange(0, 10)
y = [np.math.pow(1/2, c) for c in x]

plt.plot(x, y)
plt.show()
```

```python
# divergierende Reihe

x = np.arange(0, 10)
y = [np.math.pow(3/2, c) for c in x]

plt.plot(x, y)
plt.show()
```

### Numerische Integration

Beispiel einfachste Variante (mit Rechtecken und Intervalle):



```python
# Nicht stetige Funktion da 0 undefiniert ist

x = [c for c in np.arange(-10, 11) if c != 0]
y = [(c * -1 if c < 0 else c)/c for c in x]

plt.plot(x, y)
plt.show()
```

### Stetigkeit Fallunterscheidung



```python
# Oszilierende Sinuskurve
# nichtstetig
x = [c for c in np.arange(-10, 11, 0.1) if c != 0]
y = [np.math.sin(1/c) for c in x]

plt.plot(x, y)
plt.show()
```

```python
# Stetige Kurve
x = [c for c in np.arange(-10, 11, 0.1) if c != 0]
y = [c*np.math.sin(1/c) for c in x]

plt.plot(x, y)
plt.show()
```

### Linksseitige und rechtsseitige Grenzwerte

```python
xp = np.linspace(0.05, np.pi-0.05)
xn = -xp

fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.set_aspect(1)
ax.set_xlim((-3,3))
ax.set_ylim((-3,3))
ax.plot([-3,3], [0,0], 'k')
ax.plot([0,0], [-3,3], 'k')

ax.plot(xp, 1 / np.tan(xp), 'r')
ax.plot(xn, 1 / np.tan(xn), 'r')

ax.set_title("Graph von $\cot$")
#Berechnung der Limites bei x=0 mit Sympy:

x = sympy.Symbol('x')
print(sympy.limit(1/sympy.tan(x), x, 0, '+'))
print(sympy.limit(1/sympy.tan(x), x, 0, '-'))
```

### Graphische Darstellung und Limites von 1x:

```python
xp = np.linspace(0.05, 3, 100)
xn = -xp

fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.set_aspect(1)

ax.set_xlim((-3,3))
ax.set_ylim((-3,3))
ax.plot([-3,3], [0,0], 'k')
ax.plot([0,0], [-3,3], 'k')

ax.plot(xp, 1 / xp, 'r')
ax.plot(xn, 1 / xn, 'r')

ax.set_title("Graph von $1/x$")
# Berechnung der Limites bei x=0 mit Sympy:

print(sympy.limit(1/x, x, 0, '+'))
print(sympy.limit(1/x, x, 0, '-'))
```

### Graphische Darstellung und Limites von 1e−1/x:

```python
xp = np.linspace(0.05, 3, 300)
xn = -xp

fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.set_aspect(1)
ax.set_xlim((-3,3))
ax.set_ylim((0,6))
ax.plot([-3,3], [0,0], 'k')
ax.plot([0,0], [0,6], 'k')

ax.plot(xp, 1 / np.exp(-1/xp), 'r')
ax.plot(xn, 1 / np.exp(-1/xn), 'r')


ax.set_title("Graph von $g$")
#Berechnung der Limites bei x=0 mit Sympy:
print(sympy.limit(1/sympy.exp(-1/x), x, 0, '+'))
print(sympy.limit(1/sympy.exp(-1/x), x, 0, '-'))
```

Graphische Darstellung und Limites von 1e−1/x:

```python
xp = np.linspace(0.05, 3, 300)
xn = -xp

fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.set_aspect(1)
ax.set_xlim((-3,3))
ax.set_ylim((-0.5,1.5))
ax.plot([-3,3], [0,0], 'k')
ax.plot([0,0], [-0.5,1.5], 'k')

ax.plot(xp, 1 / (1+np.exp(-1/xp)), 'r')
ax.plot(xn, 1 / (1+np.exp(-1/xn)), 'r')

ax.set_title("Graph von $h$")
#Berechnung der Limites bei x=0 mit Sympy:
print(sympy.limit(1/(1+sympy.exp(-1/x)), x, 0, '+'))
print(sympy.limit(1/(1+sympy.exp(-1/x)), x, 0, '-'))
```

### Erdbeschleunigung g bis 10 Sekunden

```python
g = 9.81
x = np.arange(0, 10, 0.1)
y = [g*0.5*math.pow(c, 2) for c in x]
y = [343.2 if c > 343.2 else c for c in y]

plt.plot(x, y)
plt.show()
```

### Konvergenzbedingungen



```python
x = np.arange(0.1, 10, 0.1)
# Konvergent
y = [1/math.pow(1.1, c) for c in x]
# Divergent
y2 = [1/math.pow(0.9, c) for c in x]

plt.plot(x, y)
plt.plot(x, y2)
plt.show()
```
