```python
import matplotlib.pyplot as plt
import numpy as np

def expGrowth(day, factor):
    if day == 1:
        return factor
    return expGrowth(day - 1, factor) * factor
    pass

x = np.arange(1, 90)
y1 = [expGrowth(d, 1.03) for d in x]
y2 = [expGrowth(d, 1.04) for d in x]
y3 = [expGrowth(d, 1.05) for d in x]

plt.plot(x, y1, label='BEST')
plt.plot(x, y2, label='NORMAL')
plt.plot(x, y3, label='WORST')
plt.legend()
plt.show()
```
