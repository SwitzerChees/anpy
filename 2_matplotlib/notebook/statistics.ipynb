```r
x = rnorm(10000)

mean(x) #empirischer Mittelwert
```

#### Mit 99% Wahrscheinlichkeit liegt das Resulat zwischen bis

```r
n = 1000
sigma = 1 #Erwartungswert Median oder Mittelwert

mean(x)
# Mittelwert + z * Standardabweichung / Wurzel n
mean(x) + 2.8 * sigma / sqrt(n) #99.5% Quanitla. = 2.8
mean(x) - 2.8 * sigma / sqrt(n)
```

#### Sigma nicht definiert (t-Verteilung ein muss), ab n > 200 Normalverteilung

```r
mean(x)

# Mittelwert + z * Standardabweichung / Wurzel n
mean(x) + 2.2622 * sd(x) / sqrt(n) 
mean(x) - 2.2622 * sd(x) / sqrt(n)
```
